COMBINAR PARQUETS YELLOW Y PASARLOS A CSV


```python
import os
import pandas as pd

# Directorio que contiene los archivos Parquet (ruta relativa)
directory = 'YELLOW_2018'

# Obtener la lista de archivos Parquet en el directorio
parquet_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.parquet')]

# Leer y combinar los archivos Parquet en un solo DataFrame
dfs = []
for file in parquet_files:
    df = pd.read_parquet(file)
    dfs.append(df)

# Combinar todos los DataFrames en uno solo
combined_df_yellow_2018 = pd.concat(dfs, ignore_index=True)
```

```python
# Verificar si alguna columna tiene todos los valores nulos en combined_df_yellow_2018
columnas_con_valores_nulos = combined_df_yellow_2018.columns[combined_df_yellow_2018.isnull().all()]

# Mostrar las columnas con todos los valores nulos
print("Columnas con todos los valores nulos:")
print(columnas_con_valores_nulos)
```

```python
# Supongamos que tienes un DataFrame llamado combined_df_yellow_2018

columnas_a_eliminar = ['VendorID', 'RatecodeID', 'store_and_fwd_flag']

# Eliminar las columnas especificadas
combined_df_yellow_2018 = combined_df_yellow_2018.drop(columns=columnas_a_eliminar)
```
